### 3. Crea una gráfica que contenga dos líneas de diferente color con una leyenda apropiada a cada línea

```python
x = np.arange(20)
y = np.sin(x)
x1=np.arange(20)
y1=np.cos(x1)
plt.plot(y,label="Seno de 0 - 20")
plt.plot(y1,label="coseno de 0 - 20")
plt.legend(loc="best")
plt.show()
```

### 4. Crea una gráfica que contenga dos líneas de diferente color, diferente grosor, y leyenda apropiada para cada línea
