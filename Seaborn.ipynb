```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("pokemon.csv")
```

```python
df.head()
```

```python
len(df)
```

## Funciones de graficado de Seaborn

```python
# funcion lmplot() con 3 argumentos,( x, y ,de donde salen los datos)
sns.lmplot(x="Attack",y="Defense",data=df)
plt.show()
```

```python
sns.lmplot(x="Attack",y="Defense",data=df,fit_reg=False)
plt.show()
```

```python
sns.set_style("darkgrid") # temas de seaborn
sns.lmplot(x="Attack",y="Defense",data=df,fit_reg=False,hue="Stage")
plt.show()
```

## Alteraciones con Matplotlib

```python
sns.lmplot(x="Attack",y="Defense",data=df,fit_reg=False,hue="Stage")
plt.xlim(40)
plt.ylim(20)
plt.title("Ataque y defensa de un pokemon")
plt.show()
```

## El rol de pandas

```python
plt.figure(figsize=(12,9))
sns.boxplot(data=df)
```

```python
df2=df.drop(["#","Total","Stage","Legendary"],axis=1)
df2.head()
```

```python
plt.figure(figsize=(12,9))
sns.boxplot(data=df2)
```

```python
plt.figure(figsize=(14,8))
sns.violinplot(data=df2)
```

```python
# Paleta de colres
colores_tipo_pokemon = ['#78C850',  # Grass
'#F08030',  # Fire
'#6890F0',  # Water
'#A8B820',  # Bug
'#A8A878',  # Normal
'#A040A0',  # Poison
'#F8D030',  # Electric
'#E0C068',  # Ground
'#EE99AC',  # Fairy
'#C03028',  # Fighting
'#F85888',  # Psychic
'#B8A038',  # Rock
'#705898',  # Ghost
'#98D8D8',  # Ice
'#7038F8',  # Dragon
]
```

```python
plt.figure(figsize=(14,8))
sns.violinplot(x="Type 1", y="Attack", data=df2, palette=colores_tipo_pokemon)
```

```python
plt.figure(figsize=(14,8))
sns.swarmplot(x='Type 1', y='Attack', data=df, palette=colores_tipo_pokemon)
```

## Combinar gráficas

```python
plt.figure(figsize=(12,8))
 
sns.violinplot(x='Type 1',y='Attack', data=df, 
               inner=None, # quitar barras que están adentro del violín
               palette=colores_tipo_pokemon)
 
sns.swarmplot(x='Type 1', 
              y='Attack', 
              data=df, 
              color='k', # puntos negros
              alpha=0.7) # puntos un poco transparentes
plt.title('Ataque por tipo de Pokemon')
```

```python
mat_corr =df2.corr() # matriz de correlación
mat_corr
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(mat_corr)
```
