## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25])
```

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25])
plt.title("Primera gr√°fica")
plt.xlabel("Eriqueta eje x")
plt.ylabel("Etiqueta eje y")
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.plot([1,2,3,4,5],[1,4,9,16,25])
plt.title("Primera gr√°fica")
plt.xlabel("Eriqueta eje x")
plt.ylabel("Etiqueta eje y")
plt.show()
```

```python
## plot recibe 3 argumentos "forzosos", (2) ejes y estilo (go, go-,r*,r*-)
plt.plot([1,2,3,4],[1,4,9,16],"r*-")
plt.title("usando 3er argumento")
plt.xlabel("Eriqueta eje x")
plt.ylabel("Etiqueta eje y")
plt.show()
```

```python
x=np.arange(1,5)
y=x**3
print(x)
print(y)
```

```python
plt.plot([1,2,3,4],[1,4,9,16],"r*-",x, y,"g*-")
```
